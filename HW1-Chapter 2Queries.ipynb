\-- Propositon:

\-- All of the employees that assisted customer 71

\-- within a calendar year on more than one occasion.

```sql
USE Northwinds2022TSQLV7;

SELECT [EmployeeId], YEAR([OrderDate]) AS orderyear, COUNT(*) AS numorders
FROM [Northwinds2022TSQLV7].[Sales].[Order]
WHERE [CustomerId] = 71
GROUP BY [EmployeeId], YEAR([OrderDate])
HAVING COUNT(*) > 1
ORDER BY [EmployeeId], YEAR([OrderDate]);

USE TSQLV4;

SELECT empid, YEAR(orderdate) AS orderyear, COUNT(*) AS numorders
FROM Sales.Orders
WHERE custid = 71
GROUP BY empid, YEAR(orderdate)
HAVING COUNT(*) > 1
ORDER BY empid, orderyear;
```

\-- Proposition

\-- All of the employees that helped customer 71 each order year 

\-- showing the total freights for the number of orders.

```sql
USE TSQLV4;
SELECT
  empid,
  YEAR(orderdate) AS orderyear,
  SUM(freight) AS totalfreight,
  COUNT(*) AS numorders
FROM Sales.Orders
WHERE custid = 71
GROUP BY empid, YEAR(orderdate);

USE Northwinds2022TSQLV7;
SELECT
  [EmployeeId],
  YEAR([OrderDate]) AS orderyear,
  SUM([Freight]) AS totalfreight,
  COUNT(*) AS numorders
FROM [Northwinds2022TSQLV7].[Sales].[Order]
WHERE [CustomerId] = 71
GROUP BY [EmployeeId], YEAR([OrderDate]);
```

\-- Proposition:

\-- All employees who helped the total number of different customers 

\-- per year.

```sql
USE TSQLV4;
SELECT 
  empid, 
  YEAR(orderdate) AS orderyear, 
  COUNT(DISTINCT custid) AS numcusts
FROM Sales.Orders
GROUP BY empid, YEAR(orderdate);

USE Northwinds2022TSQLV7;
SELECT 
  [EmployeeId], 
  YEAR([OrderDate]) AS orderyear, 
  COUNT(DISTINCT [CustomerId]) AS numcusts
FROM [Northwinds2022TSQLV7].[Sales].[Order]
GROUP BY [EmployeeId], YEAR([OrderDate]);
```
